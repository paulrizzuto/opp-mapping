### Calculate the Distance Between Remaining Cities and Universities

```python
physical_locations = pd.read_excel("Raw/Physical_Campuses_geo.xlsx")
physical_locations.count()
```

```python
beyond_google = pd.DataFrame(columns=['City', 'State', 'City Lat', 'City Lng',
                                      'University', 'Uni CityState', 'Uni Lat', 'Uni Lng', 'Distance'])
```

```python
border_rules = {}
csvpath = 'Raw/border-rules.csv'

with open(csvpath, newline= "") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    next(csvreader)

    for row in csvreader:
        state = row[0]
        border_rules[state] = row[1:]
        
for item in border_rules:
    border_rules[item] = [x for x in border_rules[item] if x]
```

```python
counter = 0
for index, row in physical_locations.iterrows():
    print("Checking " + row["University"])
    state_pull = row["State"]
    uni_lat = row["Uni_lat"]
    uni_lng = row["Uni_lng"]
    uni_name = row["University"]
    uni_loc = row["CityState"]
    for item in border_rules[state_pull]:
        if item is not "":
            #check against the city/state dataframe, create new dataframe of values, add values for city/state and lat/lng
            for index, row in no_affiliate.iterrows():
                if row["state"] == item:
                    city_name = row["city"]
                    state_name = row["state"]
                    city_lat = row["lat"]
                    city_lng = row["lng"]
                    city_state = row["CityState"]
                    
                    p1 = (uni_lat, uni_lng)
                    p2 = (city_lat, city_lng)
                    mi_distance = vincenty(p1, p2).miles
                    
                    beyond_google.set_value(counter, "Distance", mi_distance)
                    beyond_google.set_value(counter, "City", city_name)
                    beyond_google.set_value(counter, "State", state_name)
                    beyond_google.set_value(counter, "City Lat", city_lat)
                    beyond_google.set_value(counter, "City Lng", city_lng)
                    beyond_google.set_value(counter, "University", uni_name)
                    beyond_google.set_value(counter, "Uni CityState", uni_loc)
                    beyond_google.set_value(counter, "Uni Lat", uni_lat)
                    beyond_google.set_value(counter, "Uni Lng", uni_lng)
                    beyond_google.set_value(counter, "CityState", city_state)
                    counter = counter + 1
                else:
                    continue
            #print(item)
    #print("Next university")

#testing_distance.to_csv("matrix_cities.csv")

beyond_google.head()
```

### Incorporate Rules Again

```python
#need to add another cell for university rules if used, not being used here so skipping for now in the interest of time
```

```python
#incorporate negative rules on full frame
#import university rules csv, clean up
negative_rules = {}
csvpath = 'Raw/negative-rules.csv'

with open(csvpath, newline= "") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    next(csvreader)

    for row in csvreader:
        uni = row[0]
        negative_rules[uni] = row[1:]

for item in negative_rules:
    negative_rules[item] = [x for x in negative_rules[item] if x]
```

```python
for index, row in beyond_google.iterrows():
    if row["University"] in negative_rules:
        for item in negative_rules[row["University"]]:
            if (row["State"] == item) or (row["CityState"] == item):
                beyond_google.drop(index, inplace=True)
```
